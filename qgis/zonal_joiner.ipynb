```python
import pandas as pd
```

```python
def cleaner(df):
    df["total"]= df[df.columns[-15:]].sum(axis=1)
    
    df = df.rename(columns={"HISTO_11": "open_water", "HISTO_12":"ice_snow", "HISTO_21":"open_space","HISTO_22":"low_intensity","HISTO_23":"medium_intensity","HISTO_24":"high_intensity","HISTO_31":"barren_land","HISTO_41":"deciduous_forest","HISTO_42":"evergreen_forest","HISTO_43":"mixed_forest","HISTO_51":"dwarf_scrub","HISTO_52":"shrub_scrub","HISTO_71":"grassland","HISTO_72":"sedge","HISTO_73":'lichens',"HISTO_74":"moss","HISTO_81":"pasture_hay","HISTO_82":"cultivated_crops","HISTO_90":"woody_wetlands","HISTO_95":"emergent_wetlands"})
    
    
    #convert county FIPS to str
    cFP = df["COUNTYFP"]
    cFP = list(cFP)
    cFP = [str(i) for i in cFP]
    
    #add 0's to county FP
    new_FP = []
    for i in cFP:
        if len(i) == 3:
            new_FP.append(i)
        else:
            i = i.zfill(3)
            new_FP.append(i)
            
    #convert state FIPS to str
    sFP = df["STATEFP"]
    sFP = list(sFP)
    sFP = [str(i) for i in sFP]
    
    #join FIPS
    FIPS = [i + j for i, j in zip(sFP, new_FP)]
    FIPS = [int(i) for i in FIPS]
    df['combined_FIPS'] = FIPS

    
    #remove empty data (counties outside of lower 48)
    df = df[df.total != 0]
    
    return df





    
```

```python
zonal_2016 = cleaner(zonal_2016)
```

```python
from operator import truediv
def calc_developed(df):
    
    open_space = df['open_space']
    open_space = list(open_space)

    low_intensity = df['low_intensity']
    low_intensity = list(low_intensity)

    medium_intensity = df['medium_intensity']
    medium_intensity = list(medium_intensity)

    high_intensity = df['high_intensity']
    high_intensity = list(high_intensity)
    
    count = df['total']
    count = list(count)



    dev =  [sum(x) for x in zip(open_space, low_intensity, medium_intensity, high_intensity)]
    pctDev =  map(truediv, dev, count)
    pctDev = [x for x in pctDev]
    pctDev

    return pctDev

```

```python
def calc_suburban(df):
    
    open_space = df['open_space']
    open_space = list(open_space)

    low_intensity = df['low_intensity']
    low_intensity = list(low_intensity)

    
    count = df['total']
    count = list(count)



    dev =  [sum(x) for x in zip(open_space, low_intensity)]
    pctSub =  map(truediv, dev, count)
    pctSub = [x for x in pctSub]
    pctSub

    return pctSub

```

```python
def calc_forest(df):
    deciduous_forest = df['deciduous_forest']
    deciduous_forest = list(deciduous_forest)

    evergreen_forest = df['evergreen_forest']
    evergreen_forest = list(evergreen_forest)

    mixed_forest = df['mixed_forest']
    mixed_forest = list(mixed_forest)
    
    woody_wetlands = df['woody_wetlands']
    woody_wetlands = list(woody_wetlands)
    
    count = df['total']
    count = list(count)
    
    forest =  [sum(x) for x in zip(woody_wetlands, deciduous_forest, evergreen_forest, mixed_forest)]
    forest =  map(truediv, forest, count)
    pctForest = [x for x in forest]
    return pctForest
```

```python
def calc_undeveloped(df):
    deciduous_forest = df['deciduous_forest']
    deciduous_forest = list(deciduous_forest)

    evergreen_forest = df['evergreen_forest']
    evergreen_forest = list(evergreen_forest)

    mixed_forest = df['mixed_forest']
    mixed_forest = list(mixed_forest)
    
    woody_wetlands = df['woody_wetlands']
    woody_wetlands = list(woody_wetlands)
    
    emergent_wetlands = df['emergent_wetlands']
    emergent_wetlands = list(emergent_wetlands)
    
    grassland = df['grassland']
    grassland = list(grassland)
    
    shrub_scrub = df['shrub_scrub']
    shrub_scrub = list(shrub_scrub)
    
    count = df['total']
    count = list(count)
    
    unDev =  [sum(x) for x in zip(shrub_scrub, grassland, emergent_wetlands, woody_wetlands, deciduous_forest, evergreen_forest, mixed_forest)]
    unDev =  map(truediv, unDev, count)
    pctUndev = [x for x in unDev]
    return pctUndev
```

```python
def calc_agri(df):
    pasture_hay = df['pasture_hay']
    pasture_hay = list(pasture_hay)

    cultivated_crops = df['cultivated_crops']
    cultivated_crops = list(cultivated_crops)
    
    count = df['total']
    count = list(count)
    
    ag =  [sum(x) for x in zip(cultivated_crops, pasture_hay)]
    ag =  map(truediv, ag, count)
    pctAg = [x for x in ag]
    return pctAg
```
